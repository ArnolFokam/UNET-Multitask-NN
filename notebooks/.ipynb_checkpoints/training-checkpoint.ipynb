```python
import os
import sys
import numpy as np
import pandas as pd
import tensorflow as tf

import matplotlib.pyplot as plt

sys.path.append('../')
from utils.data_generators import DicomDataGenerator
from models.metrics.dice import soft_dice_loss
```

# Preprocessing

```python
data_path='C:\\Users\\DELL\\Desktop\\COMS4059A - Research Project\\data\\processed'
df=pd.read_csv(os.path.join(data_path, 'meta/meta_info.csv'),
              dtype={'patient_id': str,
                    'nodule_no': str,
                    'slice_no': str})

# use only non-clean scans (scans that contains at least one nodule) for training
df = df[df['is_clean'] == False]
df['is_clean'].unique()
```

```python
df.head()
```

```python
def get_paths(x):
    patient_img_path=os.path.join(data_path, 'image', 'LIDC-IDRI-' + x[0])
    patient_mask_path=os.path.join(data_path, 'mask', 'LIDC-IDRI-' + x[0])
    return [os.path.join(patient_img_path, x[1]+'.npy'), os.path.join(patient_mask_path, x[2]+'.npy')]

temp = df[['patient_id', 
    'original_image', 
    'mask_image']].values

paths = list(map(get_paths, temp))
df_paths = pd.DataFrame(paths, columns=['img_path', 'mask_path'])
df_paths.head()
```

```python
df.reset_index(drop=True, inplace=True)
df_paths.reset_index(drop=True, inplace=True)

df = pd.concat([df, df_paths], axis=1, sort=False)
df.head()
```

```python
df = df[[
    'subtlety',
    'internalStructure',
    'calcification',
    'sphericity',
    'margin',
    'lobulation',
    'spiculation',
    'texture',
    'malignancy',
    'img_path',
    'mask_path'
]]
df.head()
```

# Training

```python
"""from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
df[[
    'subtlety',
    'internalStructure',
    'calcification',
    'sphericity',
    'margin',
    'lobulation',
    'spiculation',
    'texture',
    'malignancy',
]] = scaler.fit_transform(df[[
    'subtlety',
    'internalStructure',
    'calcification',
    'sphericity',
    'margin',
    'lobulation',
    'spiculation',
    'texture',
    'malignancy',
]])"""
```

```python
from sklearn.model_selection import train_test_split

batch_size=4
image_size=64
num_batches=int(len(df)/batch_size)

with tf.device('/device:GPU:0'):
    datagen = DicomDataGenerator(df,
                             img_path_col_name='img_path',
                             mask_path_col_name='mask_path',
                             features_cols=['subtlety',
                                            'internalStructure',
                                            'calcification',
                                            'sphericity',
                                            'margin',
                                            'lobulation',
                                            'spiculation',
                                            'texture'],
                                           #'malignancy'],
                             batch_size=batch_size,
                             target_size=(image_size, image_size, 1)
                            )


    traingen, valigen = train_test_split(datagen, test_size=0.33)
```

```python
from models.pnsamp_2d import PNSAMP_2D

model = PNSAMP_2D(num_attributes=8, input_size=(image_size, image_size, 1))
model.summary()
```

```python
from keras_unet_collection import models

model = models.unet_2d((image_size, image_size, 1), [64, 128, 256, 512, 1024], n_labels=1,
                      stack_num_down=2, stack_num_up=1,
                      activation='GELU', output_activation='Sigmoid', 
                      batch_norm=True, pool='max', unpool='nearest', name='unet')

model.summary()
```

```python
# Instantiate an optimizer.
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

# Instantiate a loss function.
loss_fn1 = tf.keras.losses.BinaryCrossentropy() #dice_loss
# loss_fn2 = tf.keras.losses.SparseCategoricalCrossentropy()
```

```python
epochs = 5

with tf.device('/device:GPU:0'):
    for epoch in range(epochs):
        print("\nStart of epoch %d" % (epoch,))

        # Iterate over the batches of the dataset.
        for step, (x_batch_train, mask_batch_train, feats_batch_train, mal_batch_train) in enumerate(traingen):
            # Open a GradientTape to record the operations run
            # during the forward pass, which enables auto-differentiation.
            with tf.GradientTape(persistent=True) as tape:

                # Run the forward pass of the layer.
                # The operations that the layer applies
                # to its inputs are going to be recorded
                # on the GradientTape.
                segmentation_logits = model(x_batch_train, training=True)  # Logits for this minibatch

                # Compute the loss value for this minibatch.
                loss_value1 = loss_fn1(mask_batch_train, segmentation_logits)
                # loss_value2 = loss_fn2(y_batch_train, multi_classification_logits)
                # loss_value3 = loss_fn2(np.expand_dims(mal_batch_train, axis=-1), classification_logits)
                print(loss_value1)

            """MULTI-CLASSIFICATOIN"""
            # Use the gradient tape to automatically retrieve
            # the gradients of the trainable variables with respect to the loss.
            grads = tape.gradient(loss_value1, model.trainable_weights)

            # Run one step of gradient descent by updating
            # the value of the variables to minimize the loss.
            optimizer.apply_gradients(zip(grads, model.trainable_weights))
```
